## 9. Budowa modeli regresji liniowej

### 9.1 Model 1 - pierwszy zestaw zmiennych

```python
model_1 = smf.ols(
    "systolic ~ age + smoker + fastfood + diabetes",
    data=df
).fit()

model_1.summary()
```


Pierwszy model regresji liniowej uwzględnia wyłącznie zmienne związane ze stylem
życia oraz stanem zdrowia pacjentów. Model ten stanowi punkt odniesienia
dla dalszej analizy i pozwala ocenić, w jakim stopniu czynniki pozabiometryczne
wyjaśniają zmienność skurczowego ciśnienia krwi.


### 9.2 Model 2 - drugi zestaw zmiennych

```python
model_2 = smf.ols(
    "systolic ~ age + smoker + fastfood + diabetes + bmi + waist",
    data=df
).fit()

model_2.summary()
```

Drugi model regresji został rozszerzony o zmienne antropometryczne, takie jak
wskaźnik BMI oraz obwód talii, które zgodnie z wcześniejszą analizą EDA wykazywały
silny związek ze skurczowym ciśnieniem krwi. Celem budowy modelu rozszerzonego jest sprawdzenie, czy uwzględnienie cech
budowy ciała prowadzi do poprawy jakości dopasowania modelu.


### 9.3 Wstępna ocena jakości modeli

```python
model_1.rsquared_adj, model_2.rsquared_adj
```

Wstępna analiza współczynnika determinacji skorygowanej sugeruje, że model rozszerzony charakteryzuje się lepszym dopasowaniem do danych, co wskazuje na istotny wpływ zmiennych antropometrycznych na poziom skurczowego ciśnienia krwi.


## 10. Porównanie modeli

### 10.1 Porównanie współczynnika determinacji skorygowanej 

```python
comparison = pd.DataFrame({
    "Model": ["Model 1", "Model 2"],
    "Adj_R2": [model_1.rsquared_adj, model_2.rsquared_adj]
})

comparison
```

Porównanie modeli regresji liniowej przeprowadzono na podstawie współczynnika determinacji skorygowanej, który uwzględnia zarówno stopień dopasowania modelu do danych, jak i liczbę zastosowanych predyktorów. Analiza wykazała, że model rozszerzony (Model 2) charakteryzuje się wyższą wartością
współczynnika R² skorygowanego w porównaniu z modelem podstawowym (Model 1). Oznacza to, że uwzględnienie zmiennych antropometrycznych prowadzi do poprawy jakości dopasowania modelu, mimo zwiększenia jego złożoności.


## 11. Wybór najlepszego modelu

Wybór najlepszego modelu regresji dokonano na podstawie współczynnika determinacji skorygowanej, który pozwala na ocenę jakości dopasowania modelu z uwzględnieniem liczby zastosowanych zmiennych objaśniających.

Na podstawie przeprowadzonego porównania modeli stwierdzono, że model rozszerzony (Model 2) charakteryzuje się wyższą wartością współczynnika R² skorygowanego w porównaniu z modelem podstawowym. W związku z tym Model 2 został wybrany jako model najlepiej opisujący zależność pomiędzy skurczowym ciśnieniem krwi a rozpatrywanymi zmiennymi objaśniającymi.

Choć różnica wartości współczynnika determinacji skorygowanej pomiędzy porównywanymi modelami nie jest duża, uwzględnienie zmiennych antropometrycznych w Modelu 2 pozwala na lepsze wyjaśnienie zmienności  skurczowego ciśnienia krwi bez nadmiernego zwiększenia złożoności modelu.



## 12. Postać wybranego modelu regresji

### 12.1. Współczynniki wybranego modelu

```python
model_2.params
```

### 12.2 Postać analityczna modelu

Postać wybranego modelu regresji liniowej wielorakiej można zapisać w następującej postaci:

$$
\hat{y} = \beta_0 
+ \beta_1 \cdot \text{age}
+ \beta_2 \cdot \text{smoker}
+ \beta_3 \cdot \text{fastfood}
+ \beta_4 \cdot \text{diabetes}
+ \beta_5 \cdot \text{bmi}
+ \beta_6 \cdot \text{waist}
$$

gdzie:
- $\hat{y}$ oznacza przewidywaną wartość skurczowego ciśnienia krwi,
- $\beta_0$ jest wyrazem wolnym,
- $\beta_1, \ldots, \beta_6$ są współczynnikami regresji odpowiadającymi
  poszczególnym zmiennym objaśniającym.


### 12.3 Postać numeryczna modelu

## 13. Istotność współczynników i ich interpretacja

### 13.1 Istotność statystyczna współczynników

```python
model_2.summary()
```

Istotność statystyczną współczynników regresji oceniono na podstawie wartości p-value, przyjmując poziom istotności α = 0.05.

Na podstawie wyników estymacji stwierdzono, że istotny statystycznie wpływ na skurczowe ciśnienie krwi mają zmienne `age` oraz `waist`, dla których wartości p-value są mniejsze od 0.05.

Pozostałe zmienne objaśniające, tj. `smoker`, `fastfood`, `diabetes` oraz `bmi`, nie wykazują istotności statystycznej na przyjętym poziomie istotności, co oznacza brak podstaw do odrzucenia hipotezy zerow


### 13.2 Interpretacja współczynników regresji

Interpretacja współczynników regresji została przeprowadzona przy założeniu niezmienności pozostałych zmiennych (ceteris paribus).

Współczynnik przy zmiennej `age` ma wartość dodatnią i jest istotny statystycznie, co oznacza, że wraz ze wzrostem wieku pacjenta o jeden rok przewidywana wartość skurczowego ciśnienia krwi wzrasta średnio o około 0.41 mmHg, przy pozostałych zmiennych stałych.

Zmienna `waist` również wykazuje istotny dodatni wpływ na poziom ciśnienia skurczowego. Oznacza to, że zwiększenie obwodu talii o jeden centymetr wiąże się ze wzrostem przewidywanego ciśnienia skurczowego średnio o około 0.14 mmHg, przy pozostałych zmiennych niezmienionych.


### 13.3 Interpretacja zmiennych nieistotnych statystycznie
Zmienne `smoker`, `fastfood` oraz `diabetes`, mimo że mają dodatnie oszacowania współczynników regresji, nie wykazują istotności statystycznej na poziomie α = 0.05. Oznacza to, że w obecności pozostałych zmiennych nie można jednoznacznie stwierdzić ich wpływu na poziom skurczowego ciśnienia krwi.

Współczynnik przy zmiennej `bmi` przyjmuje wartość ujemną, jednak również nie jest istotny statystycznie, co może wynikać z występowania współliniowości pomiędzy zmiennymi `bmi` i `waist`, zaobserwowanej wcześniej na etapie eksploracyjnej analizy danych.


## 14. Interpretacja współczynnika zmienności


### 14.1 Obliczanie współczynnika zmienności 
$$
CV = \frac{s_e}{\bar{y}},
$$

gdzie:
- $s_e$ oznacza odchylenie standardowe reszt modelu,
- $\bar{y}$ oznacza średnią wartość zmiennej objaśnianej.


```python
residuals = model_2.resid
cv = residuals.std() / df["systolic"].mean()
cv
```

Obliczona wartość współczynnika zmienności dla wybranego modelu regresji wynosi 12.8%. Oznacza to, że przeciętne odchylenie wartości obserwowanych skurczowego ciśnienia krwi od wartości przewidywanych przez model stanowi około 12.8% średniej wartości tej zmiennej.

Uzyskana wartość współczynnika zmienności wskazuje na umiarkowaną jakość dopasowania modelu. Model w rozsądnym stopniu opisuje zmienność skurczowego ciśnienia krwi, jednak część zmienności pozostaje niewyjaśniona, co sugeruje możliwość uwzględnienia dodatkowych czynników wpływających na poziom ciśnienia krwi w przyszłych analizach.



## 15. Analiza reszt modelu

```python
residuals = model_2.resid
fitted = model_2.fittedvalues
```

### 15.1 Normalność rozkładu reszt

```python
shapiro(residuals)
```

```python
sns.histplot(residuals, kde=True)
plt.title("Histogram reszt")
plt.xlabel("Reszty")
plt.show()
```

Normalność rozkładu reszt oceniono za pomocą testu Shapiro–Wilka oraz na podstawie histogramu reszt. Wynik testu Shapiro–Wilka wskazuje na odrzucenie hipotezy zerowej o normalności rozkładu reszt przy poziomie istotności α = 0.05 (p-value ≪ 0.05).
Histogram reszt potwierdza odchylenie od rozkładu normalnego, wskazując na prawostronną skośność oraz obecność obserwacji o większych wartościach bezwzględnych reszt. Odchylenie od normalności rozkładu reszt jest jednak zjawiskiem typowym w przypadku dużych zbiorów danych empirycznych, w szczególności danych medycznych, i nie dyskwalifikuje modelu regresji liniowej.


### 15.2 Test zerowej wartości średniej reszt

```python
ttest_1samp(residuals, 0)
```

Test t-Studenta dla jednej próby został zastosowany w celu sprawdzenia, czy średnia wartość reszt różni się istotnie od zera. Wynik testu nie daje podstaw do odrzucenia hipotezy zerowej o zerowej wartości średniej reszt (p-value ≈ 1.0).
Oznacza to, że średnia reszt jest statystycznie równa zeru, co jest zgodne z jednym z kluczowych założeń klasycznego modelu regresji liniowej.


### 15.3 Homoskedastyczność

#### 15.3.1 Wykres reszt względem wartośći dopasowynych 

```python
plt.scatter(fitted, residuals)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Wartości dopasowane")
plt.ylabel("Reszty")
plt.title("Reszty względem wartości dopasowanych")
plt.show()
```

Analiza wykresu reszt względem wartości dopasowanych sugeruje, że wariancja reszt nie jest stała w całym zakresie wartości dopasowanych. Dla wyższych wartości dopasowanych widoczny jest większy rozrzut reszt, co może wskazywać na występowanie heteroskedastyczności.


#### 15.3.2 Test Breuscha-Pagana

```python
bp_test = het_breuschpagan(residuals, model_2.model.exog)
bp_test
```

Homoskedastyczność reszt zweryfikowano formalnie za pomocą testu Breuscha–Pagana. Wynik testu wskazuje na odrzucenie hipotezy zerowej o homoskedastyczności reszt przy poziomie istotności α = 0.05 (p-value ≪ 0.05). Oznacza to, że wariancja reszt w modelu nie jest stała, co potwierdza występowanie heteroskedastyczności.


### 15.4 Autokorelacja reszt

#### 15.4.1 Obliczenie statystyki Durbin-Watsona

```python
dw_stat = durbin_watson(residuals)
dw_stat
```

Autokorelację reszt oceniono za pomocą statystyki Durbin–Watsona. Otrzymana wartość statystyki wynosi około 2.04, co jest bardzo zbliżone do wartości 2. Wynik ten wskazuje na brak istotnej autokorelacji reszt w modelu regresji liniowej, a tym samym spełnienie założenia niezależności składnika losowego.
